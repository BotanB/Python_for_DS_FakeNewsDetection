## Data Preprocessing 

### Step1: clean rows with empty value in text

```python
cleanData_true= completeDataset_True.dropna(subset=['text']) 
```

```python
prepData= cleanData_true
prepData
```

###  Step 2: Punctuation

```python
import string
string.punctuation
string.punctuation
```

```python
def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct
prepData['text_wo_punct']=prepData['text'].apply(lambda x: remove_punctuation(x))
prepData.head()
```

### Step 3: Tokenization

```python
def tokenize(text):
    split=re.split("\W+",text) 
    return split
prepData['text_wo_punct_split']=prepData['text_wo_punct'].apply(lambda x: tokenize(x.lower()))
prepData.head()
```

### Step 4: Stopwords

```python
stopword = nltk.corpus.stopwords.words('english')
print(stopword[:11])
```

```python
def remove_stopwords(text):
    text=[word for word in text if word not in stopword]
    return text
prepData['text_wo_punct_split_wo_stopwords'] = prepData['text_wo_punct_split'].apply(lambda x: remove_stopwords(x))
prepData.head()
```

### Step 4: Stemming

```python
from nltk.stem.snowball import SnowballStemmer

# Use English stemmer.
stemmer = SnowballStemmer("english")
```

```python
prepData['text_wo_punct_split_wo_stopwords_stemmed'] = prepData['text_wo_punct_split_wo_stopwords'].apply(lambda x: [stemmer.stem(y) for y in x])
```
