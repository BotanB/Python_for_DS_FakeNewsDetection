### 1. Einleitung

Heutzutage werden Nachrichten über unterschiedliche Medien an die Masse verteilt. Medien sind zum Beispiel die sozialen Medien oder auch Online-Blogs. Einerseits führen der einfache Zugang und die schnelle Verbreitung von Nachrichten dazu, dass viele Menschen die Nachrichten konsumieren. Auf der anderen Seite aber wird dadurch die schnelle Verbreitung von "Fake News" begünstigt. 

Fake News sind Nachrichten von geringer Qualität und mit absichtlich falschen Informationen. Fake News sind aber keine Neuheit. Vor des digitalen Zeitalters wurden sie hauptsächlich durch Boulevardjournalismus verbreitet. Die Verbreitung von Fake News hängt mit der Verfügbarkeit von digitalen Werkzeugen der Massenmedien zusammen.  Die weite Verbreitung von Fake News kann negative Auswirkungen auf Individuen und die Gesellschaft haben. Daher ist die Erkennung solcher Nachrichten von hoher Relevanz (Shu et al, 2017).

#### 4.2 BuzzFeed

Da alle Artikel in einem separaten XML-Format vorliegen, werden zuerst die Dateien erfasst.
Das Format innerhalb des XML sieht folgendermaßen aus:
```
<article>
    <author></author>
    <mainText></mainText>
    <orientation></orientation>
    <title></title>
    <uri></uri>
    <veracity></veracity>
</article>
```

````python
data_path = '../DataSet/BuzzFeed_Corpus_2016/articles/'

all_files = [f for f in listdir(data_path) if isfile(join(data_path, f))]
````

##### Schritt 1: Auslesen der XML-Dateien

Diese Funktion soll die XML-Datein nach den von uns festgelegten Spaltenbezeichnungen durchsuchen und diese auslesen. Hierbei lesen wir die Spalten aus:

- Title
- mainText
- uri
- veracity

````python
def get_xml_tags(path):
    doc = minidom.parse(path)
    # Title
    try:
        title = (doc.getElementsByTagName('title')[0].firstChild.data)
    except:
        title = ''
        
    # Text
    try:
        text = (doc.getElementsByTagName('mainText')[0].firstChild.data)
    except:
        text = ''
        
    # quelle
    try:
        source = (doc.getElementsByTagName('uri')[0].firstChild.data)
    except:
        source = ''
        
    # veracity
    try:
        veracity = (doc.getElementsByTagName('veracity')[0].firstChild.data)
    except:
        veracity = ''
    
    return title, text, source, veracity    
````

Im Folgenden wird die oben erstellte Funktion auf die gesamten Daten angewendet und in seperaten Listen gespeichert.

````python
l_title = []
l_text = []
l_quelle = []
l_veracity = []

for i in all_files:
    title, text, source, veracity = get_xml_tags(data_path + i)
    l_title.append(title)
    l_text.append(text)
    l_quelle.append(source)
    l_veracity.append(veracity)
````

##### Schritt 2: Erstellen des DataFrames
Hier wird aus den angelegten Listen ein DataFrame im pandas Format erstellt.

````python
df = pd.DataFrame(list(zip(l_title, l_text, l_quelle, l_veracity)), columns =['title', 'text', 'source', 'veracity']) 
df.head()
````

##### Schritt 3: Validieren der Ziel-Spalte
In den kommenden zwei Feldern wird die Zielspalte vereinheitlicht um diese später mit allen anderen Quellen vergleichen zu können. Das bedeutet:
- true = wahre Daten
- false = fake Daten

````python
def set_veracity(verasity):
    if verasity.find('false') >= 0:
        return 'false'
    elif verasity.find('true') >= 0:
        return 'true'
    else:
        return ''
````

````python
df['veracity'] = df['veracity'].apply(set_veracity)
df.head()
````

##### Schritt 3: Extrahierung der Quellenangabe aus Spalte "Text" mithilfe von Regex

Um die Quellenangabe für jede Nachricht zu erhalten, wird der Inhalt der Spalte "Text" mithilfe von regulären Ausdrücken in Text und Quelle separiert. Die Quellenangabe wird anschließend der Spalte "Source" zugewiesen.

Das Datenset sollte nur aus Nachrichten von der Quelle Reuters stammen. Nachrichten ohne Quellenangaben oder nicht richtigen Quellenangaben werden für die spätere Analyse nicht berücksichtigt. 

````python
match=0
nomatch= 0
rownumber= 0
findrows_nomatch = []
for index, row in prepData_true.iterrows():
    if "(Reuters)" in row.text: # Sucht nach "Reuters" im Text
        if " - " in row.text: # Trennung an der Stelle "-" in Quelle und Text 
            regex = r" - " 
            source = re.split(regex,row.text)[0] # Aufteilung in Quelle 
            text = re.split(regex,row.text)[1]  # Aufteilung in Text 
            row.source = source 
            row.text = text
            match= match + 1
    else:
        nomatch= nomatch +1 
        findrows_nomatch.append(rownumber)
   
    rownumber= rownumber + 1
````

#### 5.5 Satzzeichen bereinigen

Erstellung der Funktion remove_punctuation. Die Funktion remove_punctuation( ) dient zur Entfernung von Satzzeichen im Text.

````python
def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct
````

##### Schritt 1: Worte identifizieren
Zum identifizieren der Worte kommt die Bibliothek `nltk` zum Einsatz. Sie bietet die Funktion `word_tokenize` und `FreqDist` an. Beides wird benötigt um als erstes einzelne tokens zu bilden und diese dann zusammenzufasen. Das ergebnis ist ein Datenset welches die Hüfigkeit der auftretenden Worte pro Zeile beinhaltet. Das muss dann nur noch über das gesamte Datenset aggregiert werden. All das wird in der Funktion `count_words` definiert.

Im Anschluss muss die nur noch auf die Daten angewendet werden.

````python
def count_words(df):
    # tokenize
    tokenize = df.apply(word_tokenize)
    # Count
    count_single_lines = tokenize.apply(FreqDist)
    
    # aggregieren
    end = {}
    for j in count_single_lines:
        for i in j:
            try:
                end[i] = end[i] + j[i]
            except:
                end[i] = j[i]
                
    return end
````

````python
all_words_true = count_words(df_true['text'])
all_words_false = count_words(df_false['text'])
````

##### Schritt 2: Löschen von Satzzeichen und unerwünschten Zeichen

Die Funktion dient hierbei zur umwandlung der daten von einem `dict` zu einem pandas `DataFrame`. Um dies zu tun wird `key` und `count` seperat in einer Liste ausgegeben. Dies erleichtert die erstellung eines DataFrames.

````python
def get_key_count_df(liste):
    key = []
    count = []

    for i in liste:
        key.append(i[0])
        count.append(i[1])
        
    return key, count
````

#### 6.6 Unterschiede zwischen wahren und falschen Nachrichten

##### Schritt 1: Berechnen der prozentuellen Differenz zwischen "wahr" und "falsch"
Um herauszufinden wo die Differenz zwischen wahr und falsch am größten ist, muss diese ausgerechnet werden. Hierfür werden beide Datensätze (wahr und falsch) nach ihren keys (das sind die Worte) durchsucht und die prozentuale Differenz ausgerechnet. Wichtig ist dabei, dass es für beide Sets gemacht wird damit auch Worte welche nur in einem Set vokommen, berücksichtigt werden. Das Ergebnis ist ein `dict` welches die Werte für die entsprechende Worte enthält.

````python
dif_dict = {}

for i in all_words_false:
    if i in all_words_true.keys():
        if all_words_false[i] > all_words_true[i]:
            g = all_words_true[i] / all_words_false[i]
        else:
            g = all_words_false[i] / all_words_true[i]
    else:
        g = 0
    
    dif_dict[i] = g
    
for i in all_words_true:
    if i in all_words_false.keys():
        if all_words_true[i] > all_words_false[i]:
            g = all_words_false[i] / all_words_true[i]
        else:
            g = all_words_true[i] / all_words_false[i]
    else:
        g = 0
    
    dif_dict[i] = g        
````

##### Schritt 2: Level für die Visualisierung definieren
Um nicht von den Massen an Daten überflutet zu werden grenzen wir die Verteilung ein. Wir wollen nur Worte haben welche in beiden Datensätzen vorkommen (> 0 sind) und welche eine große Differenz haben (Vorkommen der Daten < 0.2)

````python
dif = {}

for i in dif_dict:
    if dif_dict[i] < 0.2 and dif_dict[i] > 0:
        dif[i] = dif_dict[i]
        
len(dif)
````

#### 6.8 N-Gramm-Ranking

Ein n-Gramm ist eine Folge von n Wörtern, wobei n eine diskrete Zahl ist, die von 1 bis unendlich reichen kann. Beim N-Gramm-Ranking werden die N-Gramme danach geordnet, wie oft sie in der Sammlung der Fake Nachrichten und True Nachrichten vorkommen.

##### Schritt 1: Berechnen der N-Gramms

Mit dem folgenden Code werden mit Hilfe der NLTK Bibliothek 2 Gramme sowie auch 3 Gramme gebildet und nach der Häufigkeit absteigend sortiert. Die häufigsten 10 n-Gramme der True-Nachrichten sowie auch der Fake-Nachrichten werden betrachtet.

````python
twograms_true=(pd.Series(nltk.ngrams(dataset_true_extend,2)).value_counts())[:10]
trigrams_true=(pd.Series(nltk.ngrams(dataset_true_extend,3)).value_counts())[:10]
twograms_false=(pd.Series(nltk.ngrams(dataset_false_extend,2)).value_counts())[:10]
trigrams_false=(pd.Series(nltk.ngrams(dataset_false_extend,3)).value_counts())[:10]
````

````python
plt.figure(figsize=(30,20))
sns.set_style('white')

plt.subplot(2,2,1)
twograms_true.sort_values().plot.barh(color='green', width=.7, figsize=(14, 10))
plt.title('10 meist vorkommenden 2 grams')
plt.ylabel('2 gram')
plt.xlabel('Häufigkeit')

plt.subplot(2,2,3)
twograms_false.sort_values().plot.barh(color='red', width=.7, figsize=(14, 10))
plt.title('10 meist vorkommenden 2 grams "fake news"')
plt.ylabel('2 gram')
plt.xlabel('Häufigkeit')
````

````python
plt.figure(figsize=(30,20))
sns.set_style('white')

plt.subplot(2,2,1)
trigrams_true.sort_values().plot.barh(color='green', width=.7, figsize=(14, 10))
plt.title('10 meist vorkommenden 3 grams')
plt.ylabel('3 gram')
plt.xlabel('Häufigkeit')

plt.subplot(2,2,3)
trigrams_false.sort_values().plot.barh(color='red', width=.7, figsize=(14, 10))
plt.title('10 meist vorkommenden 3 grams "fake news"')
plt.ylabel('3 gram')
plt.xlabel('Häufigkeit')
````

Die Häufigkeit varriiert, dies könnte auf die Verteilung im Datensatz zurückgeführt werden. Da es sich um Amerikanische Nachrichten mit der Hauptkategorie der Politik handelt, sind die Wörter nicht auffallend. Man kann hieraus zwar die Relevanz der Gesprächsthemen ableiten aber nicht ob es sich dadurch um ein Muster handelt.

Auch die genaure Betrachtung der 3-grams zeigt kein Muster auf.

#### 6.9 Assoziationsregeln

Die Assoziationsregel ist eine Technik zur Identifikation von Beziehungen zwischen verschiedenen Elementen (z.B.: Wörtern). 

Dieses Analyseverfahren wird angewendet um Rückschlüsse daraus zu ziehen, wie häufig und mit welcher Wahrscheinlichkeit bestimmte Wörter auftreten, wenn ein bestimmtes Wort/Kombination aus Wörtern in einem Artikel vorkommt.

Itemset = Wort oder auch Kombination aus Wörtern

Support-Wert = Anzahl des Vorkommens eines Itemsets in Relation zur Gesamtanzahl des Datensatzes.

#### Schritt 1: Bildung von Assoziationen

Assoziationen können mit Hilfe von unterschiedlichen Formeln berechnet werden. In diesem Projekt wird der Apriori-Algorithmus verwendet.Er dient der Auffindung sinnvoller und nützlicher Zusammenhänge in Datenbasen, die in Form von Assoziationsregeln dargestellt werden.

Um nicht von den Massen an Daten überflutet zu werden setzen wir den Schwellenwert für den Support auf minimum 0,2


##### True News:

````python
te = TransactionEncoder()# Mit dem TransactionEncoder-Objekt wird der Datensatz in ein Array-Format umwandeln
te_ary = te.fit(dataset_true).transform(dataset_true)# Über die fit-Methode lernt der TransactionEncoder die eindeutigen Bezeichnungen im Datensatz
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets_true = apriori(df, min_support=0.2, use_colnames=True)# min support = Schwellenwert für Support

print(frequent_itemsets_true.sort_values(by=['support'], ascending=False))
````

Das Ergebnis zeigt pro itemset den dazugehörigen Support-Wert. Das Wort "said" tritt in 81% der True-Nachrichten auf. Die Kombination aus den Wörtern "trump, presid, said, state, would" tritt in 20% der Nachrichten auf. Dieses Item-Set befindet sich knapp über dem Schwellenwert.

Um nach der Itemset-Länge filtern zu können, wird im folgenden Schritt diese Information als Spalte ergänzt.

````python
frequent_itemsets_true["itemset_len"] = frequent_itemsets_true["itemsets"].apply(lambda x: len(x))
print(frequent_itemsets_true)
````

Es kommt aufgrund des Schwellenwerts von 0,2 maximal eine Kombination aus 5 Wörtern vor.

Im nächsten Schritt wird die Kombination aus 2 Wörtern genauer betrachtet:

````python
association_item2= frequent_itemsets_true[ (frequent_itemsets_true['itemset_len'] == 2)] #  Filter 2er Kombi
association_item2= association_item2.sort_values(by=['support'], ascending=False) #  Sortieren nach Support
association_item2 = association_item2[:10] # Top 10 2er Kombinationen
print(association_item2)
````

Die Kombination aus `would` und `said` kommt in fast 55% der True-Nachrichten vor. Die Kombination aus `Donald` und `Trump` zeigt einen Support Wert von ca. 50 %.  Es ist ersichtlich, dass das Wort `said` in 9 der Top 10 2er-Kombinationen vorkommt. 

Im nächsten Schritt wird die Kombination aus 3 Wörtern genauer betrachtet:

````python
association_item3= frequent_itemsets_true[ (frequent_itemsets_true['itemset_len'] == 3)] #  Filter 3er Kombi
association_item3= association_item3.sort_values(by=['support'], ascending=False) #  Sortieren nach Support
association_item3 = association_item3[:10] # Top 10 3er Kombinationen
print(association_item2)
print(association_item3)
````

Auch hier ist, ersichtlich dass die 10 häufigvorkommenden Kombinationen in True-Nachrichten weitestegehend aus den Wörtern  `would`, `said` , `donald` und `trump` bestehen. Die Kombination aus `said` , `donald` und `trump` kommt in fast 43% der True-Nachrichten vor. Die Kombination aus `Donald`,`Trump` und `presid `zeigt einen Support Wert von ca. 40%.

Die genauere Betrachtung der Kombinationen auf vierer und fünfer Items zeigen keine neuen Mustern auf.

````python
#4er Kombinationen
association_item4= frequent_itemsets_true[ (frequent_itemsets_true['itemset_len'] == 4)]
association_item4= association_item4.sort_values(by=['support'], ascending=False)
association_item4 = association_item4[:10]
# 5er Kombinationen
association_item5= frequent_itemsets_true[ (frequent_itemsets_true['itemset_len'] == 5)]
association_item5= association_item5.sort_values(by=['support'], ascending=False)
association_item5 = association_item5[:10]
````
