#### 5.2 BuzzFeed-Webis Fake News Corpus 2016

Da alle Artikel in einem seperaten XML-Format vorliegen werden zuerst die Dateien erfasst.
Das Format innerhalb des XML sieht folgendermaßen aus:
```
<article>
    <author></author>
    <mainText></mainText>
    <orientation></orientation>
    <title></title>
    <uri></uri>
    <veracity></veracity>
</article>
```

````python
data_path = '../DataSet/BuzzFeed_Corpus_2016/articles/'

all_files = [f for f in listdir(data_path) if isfile(join(data_path, f))]
````

Diese Funktion soll die XML-Datein nach den von uns festgelegten Spaltenbezeichnungen durchsuchen und diese Auslesen. Hierbei lesen wir die Spalten:
- Title
- mainText
- uri
- veracity

aus.

````python
def get_xml_tags(path):
    doc = minidom.parse(path)
    # Title
    try:
        title = (doc.getElementsByTagName('title')[0].firstChild.data)
    except:
        title = ''
        
    # Text
    try:
        text = (doc.getElementsByTagName('mainText')[0].firstChild.data)
    except:
        text = ''
        
    # quelle
    try:
        source = (doc.getElementsByTagName('uri')[0].firstChild.data)
    except:
        source = ''
        
    # veracity
    try:
        veracity = (doc.getElementsByTagName('veracity')[0].firstChild.data)
    except:
        veracity = ''
    
    return title, text, source, veracity    
````

Im folgenden wird die oben angelegte Funktion auf den gesamten Daten angewendet und in seperaten Listen gespeichert.

````python
l_title = []
l_text = []
l_quelle = []
l_veracity = []

for i in all_files:
    title, text, source, veracity = get_xml_tags(data_path + i)
    l_title.append(title)
    l_text.append(text)
    l_quelle.append(source)
    l_veracity.append(veracity)
````

Hier wird aus den angelegten listen ein Data Frame im pandas Format

````python
df = pd.DataFrame(list(zip(l_title, l_text, l_quelle, l_veracity)), columns =['title', 'text', 'source', 'veracity']) 
df.head()
````

In den kommenden zwei Feldern werden wird die Zielspalte vereinheitlicht um sie später mit allen anderen Quellen vergleichen zu können. Das bedeutet:
- true = wahre Daten
- false = fake Daten

````python
def set_veracity(verasity):
    if verasity.find('false') >= 0:
        return 'false'
    elif verasity.find('true') >= 0:
        return 'true'
    else:
        return ''
````

````python
df['veracity'] = df['veracity'].apply(set_veracity)
df.head()
````

##### Step 1: Worte identifizieren
Zum identifizieren der Worte kommt die Bibliothek `nltk` zum Einsatz. Sie bietet die Funktion `word_tokenize` und `FreqDist` an. Beides wird benötigt um als erstes einzelne tokens zu bilden und diese dann zusammenzufasen. Das ergebnis ist ein Datenset welches die Hüfigkeit der auftretenden Worte pro Zeile beinhaltet. Das muss dann nur noch über das gesamte Datenset aggregiert werden. All das wird in der Funktion `count_words` definiert.

Im Anschluss muss die nur noch auf die Daten angewendet werden.

````python
def count_words(df):
    # tokenize
    tokenize = df.apply(word_tokenize)
    # Count
    count_single_lines = tokenize.apply(FreqDist)
    
    # aggregieren
    end = {}
    for j in count_single_lines:
        for i in j:
            try:
                end[i] = end[i] + j[i]
            except:
                end[i] = j[i]
                
    return end
````

````python
all_words_true = count_words(df_true['text'])
all_words_false = count_words(df_false['text'])
````

##### Step 2: Löschen von Satzzeichen und unerwünschten Zeichen
`.`  und `,` sollen zum Beispiel nicht als einzelne Worte gezählt werden, deshalb werden sie hier entfernt.

````python
remove = [',', '.', 's', ':', 'l', ';', '(', ')']

for key in remove:
    del all_words_true[key]
    del all_words_false[key]
````

##### Step 3: Die Top 40 Worte
Wir wollen uns die 40 Häufigsten Worte anschauen. Dafür müssen wir sie von dem aus dem bestehenen set extrahieren.

````python
true_top = Counter(all_words_true).most_common(40)
false_top = Counter(all_words_false).most_common(40)
````

Die Funktion dient hierbei zur umwandlung der daten von einem `dict` zu einem pandas `DataFrame`. Um dies zu tun wird `key` und `count` seperat in einer Liste ausgegeben. Dies erleichtert die erstellung eines DataFrames.

````python
def get_key_count_df(liste):
    key = []
    count = []

    for i in liste:
        key.append(i[0])
        count.append(i[1])
        
    return key, count
````

````python
# Wahr
key, count = get_key_count_df(true_top)

df_count_true = pd.DataFrame(data={'key': key, 'count': count})
df_count_true.head()
````

````python
# Fake
key, count = get_key_count_df(false_top)

df_count_false = pd.DataFrame(data={'key': key, 'count': count})
df_count_false.head()
````

#### 6.4.2 Differenz zwischen Wahr und Fake

##### Step 1: Berechnen der protzentuellen differenz zwischen wahr und falsch
Um herauszufinden wo die differenz zwischen wahr und falsch am größten ist muss diese ausgerechnet werden. Hierfür werden beine Datensätze (wahr und falsch) nach ihren keys (das sind die worte) durchsucht und die prozentuale differenz ausgerechnet. Wichtig ist dabei das es für beide Sets gemacht wird damit auch Worte welche nur in einem Set vokommen berücksichtigt werden. Das Ergebnis ist dann ein `dict` welches die Werte für die entsprechende Worte enthält.

````python
dif_dict = {}

for i in all_words_false:
    if i in all_words_true.keys():
        if all_words_false[i] > all_words_true[i]:
            g = all_words_true[i] / all_words_false[i]
        else:
            g = all_words_false[i] / all_words_true[i]
    else:
        g = 0
    
    dif_dict[i] = g
    
for i in all_words_true:
    if i in all_words_false.keys():
        if all_words_true[i] > all_words_false[i]:
            g = all_words_false[i] / all_words_true[i]
        else:
            g = all_words_true[i] / all_words_false[i]
    else:
        g = 0
    
    dif_dict[i] = g        
````

##### Step 2: Level für die Visualisierung definieren
Um nicht von den Massen an Daten überflutet zu werden grenzen wir die Verteilung ein. Wir wollen nur Worte haben welche in beiden Datensätzen vorkommen (> 0 sind) und welche eine große differenz haben (vorkommen der Daten < 0.2)

````python
dif = {}

for i in dif_dict:
    if dif_dict[i] < 0.2 and dif_dict[i] > 0:
        dif[i] = dif_dict[i]
        
len(dif)
````

##### Step 3: Auswählen der Daten
Für die Auswertung wollen wir wieder die Top 40 Worte haben. 

````python
most_dif = Counter(dif).most_common(40)
````

Da wir aber für die Auswahl der Worte alle zusammengelegt haben, ohne auf ihre zugehörigkeit zu wahren- oder falschen Nachrichten müssen wir die Liste der Worte nun aus den Original sets entnehmen und in einen neuen DataFrame eintragen

````python
dif_true = {}
dif_false = {}

for i, j in most_dif:
    dif_true[i] = all_words_true[i]
    dif_false[i] = all_words_false[i]
````

#### 6.5 N-Gramm Ranking

Ein n-Gramm ist eine Folge von n Wörtern, wobei n eine diskrete Zahl ist, die von 1 bis unendlich reichen kann. Zum Beispiel ist das Wort "Baum" ein 1-Gramm (Unigramm). Die Kombination der Wörter "mit Baum" ist ein 2-Gramm (Bigram). Ähnlich ist "Sorte eines Baums" ein 3-Gramm (Trigramm).

Beim N-Gramm-Ranking werden die N-Gramme danach geordnet, wie oft sie in der Sammlung der Fake Nachrichten und True Nachrichten vorkommen.


````python
twograms_true=(pd.Series(nltk.ngrams(words_true,2)).value_counts())[:10]
trigrams_true=(pd.Series(nltk.ngrams(words_true,3)).value_counts())[:10]
twograms_false=(pd.Series(nltk.ngrams(words_false,2)).value_counts())[:10]
trigrams_false=(pd.Series(nltk.ngrams(words_false,3)).value_counts())[:10]
````

````python
plt.figure(figsize=(30,20))
sns.set_style('white')

plt.subplot(2,2,1)
twograms_true.sort_values().plot.barh(color='green', width=.7, figsize=(14, 10))
plt.title('10 meist vorkommenden 2 grams')
plt.ylabel('2 gram')
plt.xlabel('Häufigkeit')

plt.subplot(2,2,3)
twograms_false.sort_values().plot.barh(color='red', width=.7, figsize=(14, 10))
plt.title('10 meist vorkommenden 2 grams "fake news"')
plt.ylabel('2 gram')
plt.xlabel('Häufigkeit')
````
